02 Grouping Data

```python
# Group data by product name

df.groupby('product_name')
```

```python
# Find average price per department across all orders

df.groupby('department_id').agg({'prices': ['mean']})
```

Department 19 sells cheaper items and department 12 the most expensive

```python
# Find which department sells the most items

df.groupby('department_id').agg({'department_id': ['count']})
```

I think this is a better way of seeing which department sells more on average than the "average order number" method given in the exercise.  If one person orders from department 10 in their 10th order and another person in their 1st, they have both ordered once from department 10, but the first person's order counts ten times as much towards the metric!

```python
# Min, max and mean prices per department (and variance)

df.groupby('department_id').agg({'prices': ['min', 'max', 'mean', 'var']})
```

03 Aggregating Data with transform()

```python
# Create max_order column which gives number of orders each customer has placed

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```
